## Plotting

### Line Graphs

````python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot the points using matplotlib
plt.plot(x, y) # this function is used to create a plot with a line passing through the x and y points
````

**what happens if I plot on a figure twice**

Everytime we use the plot function it tries to plot the given data on the already existing figure.

Hence if do it twice both of them gets plotted on the same plot.

The label parameter of the plot function is used to label a line plot. The labels can be viewed by writting ```plt.legend()```

````python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot once
plt.plot(x, y, label="sine") 

y = np.cos(x)
#plot twice
plt.plot(x, y, label="cosine") 
# viewing the legend
plt.legend()
````

**What if we wanted two seperate plots**

Use the ```plt.figure()``` command to create a new figure and the plot command will follow this will plot the data on the last opened figure

````python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot once
plt.figure()
plt.plot(x, y, label="sine") 

y = np.cos(x)
#plot twice
plt.figure()
plt.plot(x, y, label="cosine") 
````

## Labelling the axis, figure, changing the xticks

```plt.xlabel(), plt.ylabel()``` - to change x and y labels

```plt.xticks(), plt.yticks()``` - to change x and y ticks/markers

```plt.title``` - to change title

Also the ```plt.show()``` has to be called everytime you want to plot the figures. (required when you want to plot in .py files.)


````python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
# setting the title
plt.title("Sine curve")
# setting the x and y labels
plt.xlabel("X")
plt.ylabel("Y")

# changing the x ticks
plt.xticks(np.arange(-10,10,1))

# use to view all the figures plotted so far. Has to be called when working with py files.
plt.show()
````

## Adding color

An image is a collection of pixels. Each pixel is represented by 3 values = (R,G,B) values. Each of the correspond to intensity of Red, Blue and Green.

In matplotlib to assign a color the parameter c is used which is the normalised values of a pixel color meaning dividing the (R,G,B) values by 255.0. 

Hence the color Red = (255,0,0) will be represented by (1.0,0,0).

Here the paramter c is a list containing tuples representing the color of each data point

````python
# assigning colors

plt.scatter(x,y, c= [(1,0,0)]*len(x))
````

## Matplotlib object Hierarchy

A huge concept of matplotlib is the matplotlib object hierarchy.  
In the previous code cells, we have called plt.plot(x,y). This one-liner hides the fact that a plot is really a hierarchy of nested Python objects. A hierarchy here means that there is a tree-like structure of matplotlib objects underlying each plot.  
  
A Figure object is the outermost container for a matplotlib graphic, which can contain multiple Axes objects. One source of confusion is the name: an Axes actually translates into what we think of as an individual plot or graph (rather than the plural of axis, as we might expect).
  
You can think of the Figure object as a box-like container holding one or more Axes (actual plots). Below the Axes in the hierarchy are smaller objects such as tick marks, individual lines, legends, and text boxes. Almost every element of a chart is its own manipulable Python object, all the way down to the ticks (reference points on a scale) and labels:

````python
fig = plt.figure()
print(fig) # prints an empty container with 0 axes.
````

# Understanding subplots

plt.subplots(n,n) divides the matplotlib figure into nxn layout each having its own axis (plots).

To plot in a specific axis, we choose the particular index and use the plot function

````python
# diving matplotlib figure into 2x2 
fig,ax =  plt.subplots(2,2)

# defining the x
x = np.arange(1,5)

#Accessing each of the subplots through its index
ax[0][0].plot(x,x*x*x)
ax[0][0].set_title('cube')

ax[0][1].plot(x,np.sqrt(x))
ax[0][1].set_title('square root')


ax[1][0].plot(x,np.exp(x))
ax[1][0].set_title('exp')


ax[1][1].plot(x,np.log10(x))
ax[1][1].set_title('log')


plt.show()
````

````python
# Another way to use subplots


# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
````

# Animation function in Matplotlib

Matplotlib can also be used for animating plots. 

As such in jupyter notebooks animations could not been embedded inline. But in the case of .py files once the animation function is defined calling ```plt.show``` will work.

However from Matplotlib 1.5.1, it supports inline display of animations in the notebook with the ```to_html5_video method```, which converts the animation to an h264 encoded video and embeddeds it directly in the notebook.

````python
from matplotlib import animation, rc
from IPython.display import HTML
````

````python
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

# setting the x and y lim not necessary
ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

# initialising the plot element by plotting an empty list
line, = ax.plot([], [], lw=2)
````

````python
# animation function. This is called sequentially

# the animation function keeps changing the value of i everytime its called. This is used to update the value of y
# hence changing the plot over time. 
# Once the data is changed the new data is set on the line element initialised above and then returned.
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)
````

````python
# this is the animation function where we choose the figure, the animation function defined above, 
# the number of frames and the time delay between the two frames.
anim = animation.FuncAnimation(fig, animate,frames=100, interval=20)
````

````python
HTML(anim.to_html5_video())
````

## animating scatter plots

Suppose you have a scatter plot, scat = ax.scatter(...), then you can

scat.set_offsets(array) - change positions

scat.set_sizes(array) - set sizes

scat.set_array(array) - change color

````python
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

# setting the x and y lim not necessary
ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

# initialising the plot element by plotting an empty list
scat = ax.scatter([], [])
````

````python
# animation function. This is called sequentially

# the animation function keeps changing the value of i everytime its called. This is used to update the value of y
# hence changing the plot over time. 
# Once the data is changed the new data is set on the scatter element initialised above and then returned.
def animate(i):
    x = np.linspace(0, 2, 10)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    data = [x,y]
    data = np.asarray(data).T
    scat.set_offsets(data)
    return (scat,)
````

````python
anim = animation.FuncAnimation(fig, animate,frames=100, interval=20)
````

````python
HTML(anim.to_html5_video())
````
